<h2>Cultivation Dataset Analysis</h2>

```python
# Reading Cultivation dataset 

file_name = 'datafile (1).csv'
cult_data = pd.read_csv(file_loc+file_name, sep=',')
print(cult_data.info())
```

```python
print(cult_data.head())
```

```python
# Add additional columns

cult_data['Cost / Hectare'] = cult_data['Cost of Production (`/Quintal) C2'] * cult_data['Yield (Quintal/ Hectare)']
cult_data['Cultivation Cost / Hectare'] = cult_data['Cost of Cultivation (`/Hectare) A2+FL'] + cult_data['Cost of Cultivation (`/Hectare) C2']
```

```python
print(cult_data.head())
```

```python
print(cult_data['Crop'].unique())
print(cult_data['State'].unique())
```

```python
#Stacked bar to show what all Crops being cultivated in the state

temp = pd.crosstab(cult_data['State'],cult_data['Crop'])

temp.plot(kind='bar', stacked=True, figsize = (16,8))
```

```python
temp1 = pd.pivot_table(cult_data,index=["Crop"],values=["Cost / Hectare"],aggfunc=np.sum).reset_index()
temp2 = pd.pivot_table(cult_data,index=["Crop"],values=["Cultivation Cost / Hectare"],aggfunc=np.sum).reset_index()
temp3 = pd.pivot_table(cult_data,index=["Crop"],values=["Cost of Production (`/Quintal) C2"],aggfunc=np.sum).reset_index()
temp4 = pd.pivot_table(cult_data,index=["Crop"],values=["Yield (Quintal/ Hectare) "],aggfunc=np.sum).reset_index()
```

```python
# Pie chart for "per Hectare Cost of each Crop"
labels = temp1['Crop']
sizes = temp1['Cost / Hectare']

#pie, patches, texts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Per Hectare Cost of each Crop',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')
#plt.tight_layout()
plt.show()
```

```python
# Pie chart for "per Hectare Cultivation Cost of each Crop"
labels = temp2['Crop']
sizes = temp2['Cultivation Cost / Hectare']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Per Hectare Cultivation Cost',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')

plt.show()
```

```python
# Pie chart for "per Quintal Cost of Production"
labels = temp3['Crop']
sizes = temp3['Cost of Production (`/Quintal) C2']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Per Quintal Cost of Production',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')

plt.show()
```

```python
# Pie chart for "Yield (Quintal/ Hectare)"
labels = temp4['Crop']
sizes = temp4['Yield (Quintal/ Hectare) ']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=0)
plt.title('Crop Yield (Quintal/ Hectare)',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')

plt.show()
```

<h3>Observation</h3>

<b><u>Crop Yield in India</u></b>

1) Sugarcane has a highest Yield in India (82.2%) <br>
2) Moong has a lowest Yeild in India (0.437%)

<b><u>Cultivation Cost for crop</u></b>

1) Higest for sugarcane <br>
2) lowest for moong <br>
3) average for 'paddy, maze, cotton, wheat' etc.

<b><u>Production cost for crop</u></b>

1) Sugarcane has very minimum production cost. <br>
2) Moong has a highest production cost and thus least Yield among country.

```python
state1 = pd.pivot_table(cult_data,index=["State"],values=["Cost / Hectare"],aggfunc=np.sum).reset_index()
state2 = pd.pivot_table(cult_data,index=["State"],values=["Cultivation Cost / Hectare"],aggfunc=np.sum).reset_index()
state3 = pd.pivot_table(cult_data,index=["State"],values=["Cost of Production (`/Quintal) C2"],aggfunc=np.sum).reset_index()
state4 = pd.pivot_table(cult_data,index=["State"],values=["Yield (Quintal/ Hectare) "],aggfunc=np.sum).reset_index()
```

```python
# Pie chart for "per Hectare Cost in each state"
labels = state1['State']
sizes = state1['Cost / Hectare']

#pie, patches, texts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Statewise Per Hectare Cost',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')
#plt.tight_layout()
plt.show()
```

```python
# Pie chart for "per Hectare Cultivation Cost in each state"
labels = state2['State']
sizes = state2['Cultivation Cost / Hectare']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Statewise Per Hectare Cultivation Cost',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')

plt.show()
```

```python
# Pie chart for "per Quintal Cost of Production in each state"
labels = state3['State']
sizes = state3['Cost of Production (`/Quintal) C2']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Statewise Per Quintal Cost of Production',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')

plt.show()
```

```python
# Pie chart for "Yield (Quintal/ Hectare) in each State"
labels = state4['State']
sizes = state4['Yield (Quintal/ Hectare) ']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=0)
plt.title('Statewise Crop Yield (Quintal/ Hectare)',loc='right',bbox={'facecolor':'0.8', 'pad':10})
plt.axis('equal')

plt.show()
```
